ECDF PLOT

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x='tip',data=tips);
```

```python
sns.ecdfplot(x='tip',data=tips,hue='time')
```
