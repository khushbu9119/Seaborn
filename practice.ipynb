## KDE Plot

```python
cars = sns.load_dataset('mpg')
cars.head()
```

```python
cars.shape
```

```python
#univariate plot
sns.kdeplot(cars.horsepower,fill=True,color='pink')
```

```python
#bivariate plot
sns.kdeplot(x=cars.horsepower, y=cars.mpg,fill=True)
plt.figure(figsize=(3, 4))
```

## HISTPLOT

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(x=penguins.bill_length_mm,bins=20)
```

```python

sns.histplot(x=penguins.bill_length_mm,data=penguins,kde=True)
```

```python
sns.histplot(x=penguins.bill_length_mm,data=penguins,hue='species')
```

```python
sns.histplot(x=penguins.bill_length_mm,data=penguins,hue='species',
             element='poly')
```

```python
sns.histplot(x=penguins.bill_length_mm,y=penguins.bill_depth_mm,data=penguins,cbar='True')
```
